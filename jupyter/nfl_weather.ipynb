```python
weather_df = pd.read_csv("http://nflsavant.com/dump/weather_20131231.csv")
weather_df
```

```python
weather_df['date2']= pd.to_datetime(weather_df['date'])
nfl_weather_df = weather_df[(weather_df['date2'].dt.year>2008) & (weather_df['date2'].dt.year<2014)]
nfl_weather_df 
```

```python
nfl_weather_df['date_string']=nfl_weather_df['id'].map(lambda x: str(x)[:-4])
```
